```python
# creating a science vocab and then filtering out words. This vocab can be customised according to your organisation

science_vocab = set()
with open('vocab.txt', encoding="utf8") as f:
  text = f.readline()
  while text:
    science_vocab.add(text.lower().strip())
    text = f.readline()

print(science_vocab)
```

```python
# creating a final list of important and relevant terms

keywords_upd = []

for keyword in keywords:
  if len(keyword.split())>1:
    keywords_upd.append(keyword)
  else:
    if keyword in science_vocab:
      keywords_upd.append(keyword)

print(keywords_upd)
```

```python
# summary part

quick_summary = []

# if wikipedia page doesn't exist then code doesn't execute.

for keyword in keywords_upd:
  try:
    summary = wikipedia.summary(keyword, sentences = 2)
    quick_summary.append(summary)
    print(keyword)
    print(summary)
  except:
    pass

# # printing the result
# for ind_quick_summary in quick_summary:
#   print(ind_quick_summary)
```

```python
def preventDocumentBreak(document):
  tags = document.element.xpath('//w:tr')
  rows = len(tags)
  for row in range(0,rows):
    tag = tags[row]                     # Specify which <w:r> tag you want
    child = OxmlElement('w:cantSplit')  # Create arbitrary tag
    tag.append(child)                   # Append in the new tag

```

```python
# creating document for upload

from docx import Document
from docx.oxml.shared import OxmlElement 

document = Document()

document.add_heading('Revision Assistant', 0)
document.add_heading('Tags: \n'+"  |  ".join(top_labels) + "\n", level=1)

table = document.add_table(rows=0, cols=2)

# summary part
quick_summary = []

# if wikipedia page doesn't exist then code doesn't execute.
for keyword in keywords_upd:
  try:
    summary = wikipedia.summary(keyword, sentences = 2)
    quick_summary.append(summary)
    row_cells = table.add_row().cells
    p = row_cells[0].add_paragraph('')
    r = p.add_run(keyword.title())
    r.bold = True 
    l = summary.split('. ')
    for bullet in l:
      row_cells[1].add_paragraph(bullet,
                    style='List Bullet')
  except:
    pass

preventDocumentBreak(document)

document.save('demo.docx')
```
